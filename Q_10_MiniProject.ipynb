```python
import numpy as np
```

```python
# Generate marks for 10 students and 5 subjects
marks = np.random.randint(30, 101, (10, 5))
print("Student Marks:",marks)

# Total marks of each student
total = np.sum(marks, axis=1)
print("\nTotal Marks:",total)

# Average marks of each student
average = np.mean(marks, axis=1)
print("\nAverage Marks:",average)

# Student with highest and lowest total marks
print("\nHighest Total Marks Student Index:", np.argmax(total))
print("Lowest Total Marks Student Index:", np.argmin(total))

# Overall class mean and standard deviation
print("\nClass Mean:", np.mean(marks))
print("Class Standard Deviation:", np.std(marks))

# Top 3 students based on total marks
top3 = np.argsort(total)[-3:]

print("\nTop 3 Students Marks:")
print(marks[top3])

# Reshape array (for demonstration)
reshaped = marks.reshape(5, 10)

print("\nReshaped Array (5 x 10):")
print(reshaped)
```
