```python
import numpy as np
```

```python
# Create 1D array of 20 random numbers
arr = np.random.randint(1, 51, 20)
print("Array:")
print(arr)

# Print minimum value
print("\nMinimum value:", arr.min())

# Print index of minimum value
print("\nIndex of minimum value:", arr.argmin())

# Print maximum value
print("\nMaximum value:", arr.max())

# Print index of maximum value
print("\nIndex of maximum value:", arr.argmax())

# Print sum
print("\nSum:", arr.sum())

# Print mean
print("\nMean:", arr.mean())

# Print standard deviation
print("\nStandard Deviation:", arr.std())
```
