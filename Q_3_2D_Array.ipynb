```python
import numpy as np
```

```python
# Create 4x5 random matrix
arr = np.random.randint(20, 81, (4, 5))

print("Matrix:")
print(arr)

# Print minimum value
print("\nMinimum Value:", arr.min())

# Print maximum value
print("\nMaximum Value:", arr.max())

# Print sum
print("\nSum:", arr.sum())

# Print mean
print("\nMean:", arr.mean())

# Print standard deviation
print("\nStandard Deviation:", arr.std())

print()

# Print row-wise sum
print("\nSum of Each Row:")
print(arr.sum(axis=1))

print()

# Print column-wise sum
print("\nSum of Each Column:")
print(arr.sum(axis=0))
```
