```python
import numpy as np
```

```python
# Create 1D array
arr = np.arange(1, 25)

# Reshape into 4x6
arr1 = arr.reshape(4, 6)

# Reshape into 3x8
arr2 = arr.reshape(3, 8)

# Reshape into 2x3x4
arr3 = arr.reshape(2, 3, 4)

# Print original array
print("Original Array:")
print(arr)

print()

# Print 4x6 array
print("4x6 Array:")
print(arr1)
print("Shape:", arr1.shape)

print()

# Print 3x8 array
print("3x8 Array:")
print(arr2)
print("Shape:", arr2.shape)

print()

# Print 3D array
print("2x3x4 Array:")
print(arr3)
print("Shape:", arr3.shape)
```
