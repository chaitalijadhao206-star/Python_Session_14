```python
import numpy as np
```

```python
# Create first array
a = np.array([10, 20, 30, 40])
print("Array A:",a)

# Create second array
b = np.array([1, 2, 3, 4])
print("\nArray B:",b)

# Addition
add = a + b
print("\nAddition:",add)

# Subtraction
sub = a - b
print("\nSubtraction:",sub)

# Multiplication
mul = a * b
print("\nMultiplication:",mul)

# Division
div = a / b
print("\nDivision:",div)

# Power
power = a ** b
print("\nPower:",power)

# Modulo
mod = a % b
print("\nModulo:",mod)
```
