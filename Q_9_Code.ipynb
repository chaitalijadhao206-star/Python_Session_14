```python
import numpy as np
```

```python
# Generate a 6x6 matrix of random numbers
arr = np.random.randn(6, 6)

# Print properties
print("Shape:", arr.shape)
print("Size:", arr.size)
print("Data Type:", arr.dtype)

# Find index of maximum and minimum value
print("Index of Maximum Value:", np.unravel_index(arr.argmax(), arr.shape))
print("Index of Minimum Value:", np.unravel_index(arr.argmin(), arr.shape))

# Extract top-left 3x3 submatrix
print("\nTop-left 3x3 Submatrix:")
print(arr[:3, :3])

# Replace negative numbers with absolute value
arr = np.abs(arr)

# Print mean of modified matrix
print("\nMean of Modified Matrix:", arr.mean())
```
